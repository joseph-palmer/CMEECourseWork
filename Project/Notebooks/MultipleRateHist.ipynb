```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib


matplotlib.rcParams.update({'font.size': 10})
fig, ax = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(6, 8))

n = 100000
d1 = np.random.exponential(1.8, n)
d2 = np.random.exponential(4.4, n)
d3 = np.random.exponential(7.5, n)
d4 = np.random.exponential(12.5, n)
d5 = np.random.exponential(16.7, n)

dall = np.concatenate([d1, d2, d3, d4, d5])

ax[0, 0].hist(d1, color="blue", alpha=0.2, label='$\lambda = 1.8$')
ax[0, 0].hist(d2, color="red", alpha=0.2, label='$\lambda = 4.4$')
ax[0, 0].legend(loc='upper right')
ax[0, 0].set_xlim([0, 100])

ax[0, 1].hist(d1, color="blue", alpha=0.2, label='$\lambda = 1.8$')
ax[0, 1].hist(d2, color="red", alpha=0.2, label='$\lambda = 4.4$')
ax[0, 1].hist(d3, color="green", alpha=0.2, label='$\lambda = 7.5$')
ax[0, 1].legend(loc='upper right')
ax[0, 1].set_xlim([0, 100])

ax[1, 0].hist(d1, color="blue", alpha=0.2, label='$\lambda = 1.8$')
ax[1, 0].hist(d2, color="red", alpha=0.2, label='$\lambda = 4.4$')
ax[1, 0].hist(d3, color="green", alpha=0.2, label='$\lambda = 7.5$')
ax[1, 0].hist(d4, color="black", alpha=0.2, label='$\lambda = 12.5$')
ax[1, 0].legend(loc='upper right')
ax[1, 0].set_xlim([0, 100])

ax[1, 1].hist(d1, color="blue", alpha=0.2, label='$\lambda = 1.8$')
ax[1, 1].hist(d2, color="red", alpha=0.2, label='$\lambda = 4.4$')
ax[1, 1].hist(d3, color="green", alpha=0.2, label='$\lambda = 7.5$')
ax[1, 1].hist(d4, color="black", alpha=0.2, label='$\lambda = 12.5$')
ax[1, 1].hist(d5, color="yellow", alpha=0.2, label='$\lambda = 16.7$')
ax[1, 1].legend(loc='upper right')
ax[1, 1].set_xlim([0, 100])

# put labels on
fig.text(0.5, 0.05, '$x$', ha='center')
fig.text(0, 0.5, 'Frequency', va='center', rotation='vertical')
fig.text(0.06, 0.88, 'A', fontweight='bold')
fig.text(0.06, 0.49, 'C', fontweight='bold')
fig.text(0.51, 0.88, 'B', fontweight='bold')
fig.text(0.51, 0.49, 'D', fontweight='bold')
fig.wspace = 2.5

# save figure
#path = "../Results/Plots/SyntheticData_Hist_5.pdf"
#fig.savefig(path)
```
