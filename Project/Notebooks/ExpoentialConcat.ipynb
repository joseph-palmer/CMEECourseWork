```python
# import tools #
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sp
from scipy.optimize import minimize
plt.style.use("seaborn-white")

# add code path to environment #
import sys
sys.path.insert(1, "../Code/")

# import package for sum of exponential fitting #
import config
from sumexp import *

def ll_exp(lamb, x):
    # Likelihood function for exponential distribution
    eq = lamb * np.exp(-lamb * x)
    ll = np.sum(np.log(eq))
    return ll

def two_exp(lamb1, lamb2, x):
    eq = 0.43*lamb1*np.exp(-lamb1*x) + 0.57*lamb2*np.exp(-lamb2*x)
    ll = np.sum(np.log(eq))
    return ll

def exp_eq(lamb, x):
    # Equation for exponential distribution [CCDF]
    eq = 1-np.exp(-lamb * x)
    return 1-eq
```

```python
# get data from exponential
expdf = np.random.exponential(1/1.8, 1000)

# define range for sampling
rts = np.arange(0.1, 8, 0.1)
res = np.zeros([len(rts), 2])

for i in range(len(rts)):
    res[i, 0] = ll_exp(rts[i], expdf)
    res[i, 1] = rts[i]

# show likelihood curve
plt.plot(res[:,1], res[:,0])
min_idx = np.where(res[:,0] == max(res[:,0]))
print(res[min_idx])
print(1/np.mean(expdf))
```

```python
def mle_diff(r1, r2):
    expdf = np.random.exponential(1/r1, 1000)
    expdf2 = np.random.exponential(1/r2, 1000)
    expdf_all = np.concatenate((expdf, expdf2))

    x = y = np.arange(0.1, 10, 0.1)
    ll = np.zeros((len(x), len(y)))
    for i in range(len(x)):
        for j in range(len(y)):
            ll[i][j] = two_exp(x[i], x[j], expdf_all)

    # mark highest point
    mle = np.amax(ll)
    max_ll_poss = np.where(ll==np.amax(ll))
    max_x = x[max_ll_poss[1]]
    max_y = y[max_ll_poss[0]]
    
    amle_1 = 1/np.mean(expdf)
    amle_2 = 1/np.mean(expdf2)
    mle_1 = max_y[0]
    mle_2 = max_y[1]
    print("~~~~~")
    print("AMLE1: {}\nMLE1 {}\ndiff1: {}\nALME2: {}\nMLE2: {}\ndiff2: {}".format(amle_1, mle_1, amle_1-mle_1, amle_2, mle_2, amle_2-mle_2))
    return(amle_1 - mle_1, amle_2 - mle_2)
```
